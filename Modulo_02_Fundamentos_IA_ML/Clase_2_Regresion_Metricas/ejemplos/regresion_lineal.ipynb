EJEMPLO DE REGRESIÓN LINEAL SIMPLE

1. FASE DE RECOLECCION DE DATOS

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Función f(x) = 0.2*x + 1.25 + 0.2*Ruido_Gaussiano
def f(x):
    np.random.seed(42)
    y = 0.2*x + 1.25 + 0.2*np.random.randn(x.shape[0])
    return y
```

```python
# Generamos valores x de 0 a 20 en intervalos de 0.3
x = np.arange(0,25,0.6)
```

```python
# Asignamos la función que hemos generado a la variable dependiente y 
y = f(x)
```

```python
# Hacemos un gráfico de los datos que hemos generado
plt.scatter(x, y, label='data', color='black')
plt.title('Datos');
```

2. FASE ENTRENAMIENTO

```python
# Importamos la clase de Regresión Lineal de scikit-learn
from sklearn.linear_model import LinearRegression
```

```python
# Creamos el objeto de LinearRegression
regression_lineal = LinearRegression()

# Entrenamos el modelo con los datos (x,y)
regression_lineal.fit(x.reshape(-1, 1), y)

# Vemos los coeficientes obtenidos
print('m = '+ str(regression_lineal.coef_) + ', b = ' + str(regression_lineal.intercept_))
```

3. FASE VALIDACION

```python
from sklearn.metrics import mean_squared_error
```

```python
prediccion_entrenamiento = regression_lineal.predict(x.reshape(-1, 1))
```

```python
mse = mean_squared_error(y_true = y, y_pred=prediccion_entrenamiento)
print(f'Error Cuadrático Medio (MSE) = {mse}')
```

```python
rmse = np.sqrt(mse)
print(f'Raiz del Error Cuadrático Medio: {rmse}')
```

```python
r2 = regression_lineal.score(x.reshape(-1,1), y)
print(f'Coeficiente de Determinación R2 = {r2}')
```

```python
plt.scatter(x, y, label='data', color='black')
plt.plot(x, prediccion_entrenamiento, color='blue', linewidth=3)
plt.show()
```
