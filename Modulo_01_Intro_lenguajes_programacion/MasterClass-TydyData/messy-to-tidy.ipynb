```python
pd.concat(
    [
        (
            pd.read_csv('data/2014-baby-names-raw.csv').assign(year=2014)
        ), 
        (
            pd.read_csv('data/2015-baby-names-raw.csv').assign(year=2015)
        )
    ]
).head()
```

```python
import glob

path = './data/'

dfs = []
for file in glob.glob(f'{path}/*baby-names-raw.csv'):
    df = pd.read_csv(file)
    df['year'] = file.split('-')[0].split('/')[-1]
    dfs.append(df)

df_baby_tidy_ex = (
    pd.concat(dfs)
    [['year', 'name', 'sex', 'rank', 'frequency']]    
    .sort_values(['year', 'frequency', 'name'], ascending=[True, False, True])
)

df_baby_tidy_ex.head()
```

```python
df_baby_tidy_ex[df_baby_tidy_ex['year'] == '2014'].nlargest(3, 'frequency')
```

```python
df_baby_tidy_ex[df_baby_tidy_ex['year'] == '2015'].nsmallest(3, 'frequency')
```

```python
# which year there are more babies with the name William?
df_baby_tidy_ex[df_baby_tidy_ex['name'] == 'William'].groupby('year').sum()
```

```python
df_baby_tidy_ex.loc[lambda df: df.name == 'William'].sort_values('frequency', ascending=False)
```
